```python
import pandas as pd
%matplotlib inline

from matplotlib import pyplot as plt
```

```python
df_data=pd.read_csv('Deals summary.csv')
names=[('Deals Option Selected','D_O'),('Location Picked','L_P'),('Recommendations viewed','R_V'),('Deal Details Loaded','D_D'),('Offer added','O_A'),('Deal Selected','D_S'),('Payment Options Viewed','P_V'),('Succesful Payment','S_P'),('Order Completed','O_C')]

x_pos_values=list(map(lambda x:x[1],names))
```

```python
df_values=[values for values in df_data.iloc[0]][1:]
print(df_values)
```

```python
x_pos=range(9)
plt.figure(figsize=(13,8))

plt.bar(x_pos,df_values,align='center',width=0.8,color='green')

plt.xticks(x_pos,x_pos_values)

plt.show()
```

```python
x_pos=range(4)
plt.bar(x_pos,df_values[2:6],align='center',width=0.8,color='orange')
plt.xticks(x_pos,list(map(lambda x:x[1],names))[2:6])

plt.show()

```

```python
#finiding drop out ratio of users at each step
drop_out=[]
drop_out_points=[]
for i in range(0,len(df_values),1):
    if len(df_values[i:i+2])==2:
        print("drop out ratio between step  {} to {} : ".format(i,i+1),df_values[i:i+2][0]-df_values[i:i+2][1])
        print('percentage drop_out {}%'.format(round(((df_values[i:i+2][0]-df_values[i:i+2][1])/df_values[i:i+2][0])*100)))
        drop_out.append(round(((df_values[i:i+2][0]-df_values[i:i+2][1])/df_values[i:i+2][0])*100))
        drop_out_points.append(df_values[i:i+2][0]-df_values[i:i+2][1])

print(sorted(drop_out,reverse=True))
```

```python
x_pos=range(8)
Step_drop_off =["{} - {}".format(list(map(lambda x:x[1],names[i:i+2]))[0],list(map(lambda x:x[1],names[i:i+2]))[1]) for i in range(0,len(names),1) if len(names[i:i+2])==2]
plt.figure(figsize=(13,8))
plt.bar(x_pos,drop_out,align='center',width=0.8)
plt.xticks(x_pos,Step_drop_off)
```

```python
import numpy as np


plt.figure(figsize=(13,8))

N = 8
oneMeans = [135, 124, 108, 65, 29, 22, 12, 3]
twoMeans = [124, 108, 65, 29, 22, 12, 3, 3]
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, oneMeans, width,align='center',yerr=drop_out_points,color='gold')
p2 = plt.bar(ind, twoMeans, width,
             bottom=oneMeans,align='center')

plt.ylabel('Scores')
plt.title('Drop_out at each step of App')
plt.xticks(ind, Step_drop_off)

plt.yticks(np.arange(0, 300, 20))



plt.show()
```
